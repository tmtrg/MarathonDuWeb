```python
resultats_df = pd.DataFrame(columns=['produit', 'mois', 'prediction'])

# boucle à travers chaque élément de la colonne "fruit et légumes"
for produit in df['Libellé'].unique():
    
    # filtrer le DataFrame pour ne conserver que les données du produit actuel
    produit_df = df[df['Libellé'] == produit].reset_index(drop=True)
    
    # renommer les colonnes pour Prophet
    produit_df = produit_df.rename(columns={'Date': 'ds', 'Moyen': 'y'})
    
    # initialiser le modèle Prophet et ajuster les données
    modele = Prophet()
    modele.fit(produit_df)
    
    # faire des prédictions pour les prochains 12 mois
    predictions = modele.make_future_dataframe(periods=12, freq='M')
    forecast = modele.predict(predictions)[-12:]
    
    # ajouter les prédictions à la DataFrame de résultats
    for i, row in forecast.iterrows():
        resultats_df = resultats_df.append({'produit': produit,
                                            'mois': row['ds'],
                                            'prediction': row['yhat']},
                                           ignore_index=True)

print(resultats_df)
```
